```python
%%manim -qh -v WARNING OpeningManim



class OpeningManim(Scene):
    def construct(self):
        title = Tex(r"Esperimento moto di caduta libera ")
        basel = Tex(r"Fabio Massimo Proietti\\ 2F", font_size=36)
        VGroup(title, basel).arrange(DOWN)
        self.play(
            Write(title),
            FadeIn(basel, shift=DOWN),
        )
        self.wait()

        transform_title = Tex("La teoria")
        transform_title.to_corner(UP + LEFT)

        self.play(
            Transform(title, transform_title),
            LaggedStart(*[FadeOut(obj, shift=DOWN) for obj in basel]),
        )

        basel = Tex(r"La teoria del moto di caduta libera si basa su quella \\ del \emph{Moto Rettilineo Uniformemente Accellerato}\\ ...con qualche differenza, riguardante l'impostazione \\ del sistema di riferimento.")
        VGroup(basel).arrange(DOWN)

        self.play(
            Write(basel),
        )
        self.wait()
                
        self.wait(5)

        self.play(
            FadeOut(basel),
        )


#######################SdR#####################################
        newtit = Tex(r"MRUA | Esempio")
        newtit.to_corner(UP + LEFT)

        self.play(
           # Create(plane, run_time=3, lag_ratio=0.1),
            Transform(title, newtit)
        )

        #ax

#######################################################################
        #1st arrow setup
        arrow1 = Arrow(1.5 * LEFT, 1.5 * RIGHT, color=GOLD)
        arrow1.shift(0.2 * DOWN)
        arw1tit = MathTex(r"a")
        arw1tit.shift(0.5 * DOWN)


        #2nd arrow setup
        arrow2 = Arrow(4.75 * LEFT, -1.5 * RIGHT, color=BLUE)
        arw2tit = MathTex(r"v_0")
        pos = 1.5
        titpos = pos + 0.5
        arw2tit.shift(titpos * DOWN, 3.15 * LEFT)
        arrow2.shift(pos * DOWN)


        #2nd arrow setup
        vf = Arrow(1.5 * LEFT, 4.75 * RIGHT, color=TEAL)
        vftit = MathTex(r"v")
        pos = 1.5
        titpos = pos + 0.5
        vftit.shift(titpos * DOWN, 1.6 * RIGHT)
        print(3.15 * RIGHT)
        vf.shift(pos * DOWN)


       ##x arrow  
        x = Arrow(5 *LEFT, 5 * RIGHT, color=RED)
        xtit = MathTex(r"x")
        xtit.shift(3.2 * DOWN)
        x.shift(2.8 * DOWN)

        self.play(
            Write(xtit),
            GrowArrow(x),
          
        )

        self.wait(0.5)

        self.play(
            Write(arw2tit),
            GrowArrow(arrow2, point_color=RED),
        )

        self.wait(0.5)

        self.play(
            Write(arw1tit),
            GrowArrow(arrow1),

        )

        self.wait(0.25)

        self.play(
            Write(vftit),
            GrowArrow(vf, run_time=0.25),
        )
        
        Lor = Tex(r"Nel caso sotto, la legge oraria del moto seguirebbe la formula \\[2pt] $ x = x_0 + v_0t +\frac 12 a t^2 $", font_size=42)
        Lor.shift(1.25*UP)

        self.play(Write(Lor))

        self.wait(3)

        a = VGroup(Lor, arw1tit, arw2tit, arrow1, arrow2, x, xtit, vf, vftit)
        self.play(
            LaggedStart(*[FadeOut(obj, shift=DOWN) for obj in a])
        )
        self.wait(3)

```
